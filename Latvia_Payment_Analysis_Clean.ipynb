```python
import pandas as pd

# 1. Load your file
df = pd.read_csv('latvia_data.csv')

# 2. Fix the messy column names to be simple
df.columns = ['Full_Date', 'Year', 'Growth_Pct', 'Value_Millions']

# 3. Fill the empty spot in the first year (2000) with a 0
df['Growth_Pct'] = df['Growth_Pct'].fillna(0)

# 4. Show the last 5 years to see the ISO 20022 era
print("--- LATVIA PAYMENT DATA (ISO 20022 ANALYSIS) ---")
print(df.tail(5))

# 5. Create a clean file for our dashboard
df.to_csv('latvia_clean_final.csv', index=False)
```
